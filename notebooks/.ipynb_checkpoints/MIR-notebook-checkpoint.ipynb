```python
#Used to define all libraries used in this code

import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

```

## Feature extraction 

We compute standard DSP/ML features such as: 
- Mel-spectogram
- MFCCs
- Spectral centroid / rolloff
- Zero-crossing rate
- Chroma features

These will serve as the numerical input to the ML moderl

```python
# Will need to iterate for each of the files inside the directory.
# Is it better to use the loop inside the plot or not
#audio_path = "../audio/01. Something\'s in the Water.wav"
audio_path = "../audio/02. Fame.wav"

# y is waveform, sr is sampling rate.
y, sr = librosa.load(audio_path)
print("Loaded:", audio_path)

# Check what is y type, and what are their features
# Check the impact of sampling rate
print("Shape:", y.shape, "Sample rate:", sr)
```

```python
fig, axs = plt.subplots(2,1, figsize=(12,6), sharex=True)

# Waveform
librosa.display.waveshow(y, sr=sr, ax=axs[0])
axs[0].set_title("Waveform - FAME")

# RMS Energy
rms = librosa.feature.rms(y=y)[0]
axs[1].plot(librosa.times_like(rms),rms)
axs[1].set_title("RMS Energy")



plt.tight_layout()
plt.show()
```

```python
# --- HARD RESET ---
plt.close('all')
fig = None   # ensure no leaked figures

### Normalize features
def normalize(x):
    return (x - x.min()) / (x.max() - x.min())

rms_n = normalize(rms)
centroid_n = normalize(centroid)
zcr_n = normalize(zcr)

t_rms = librosa.times_like(rms)
t_centroid = librosa.times_like(centroid)
t_zcr = librosa.times_like(zcr)

# --- MAIN PLOT (only one allowed) ---
fig, ax1 = plt.subplots(figsize=(12, 4))

# RMS + ZCR (left axis)
ax1.plot(t_rms, rms_n, label="RMS (norm)")
ax1.plot(t_zcr, zcr_n, label="ZCR (norm)")
ax1.set_ylabel("RMS / ZCR (normalized)")
ax1.set_xlabel("Time (s)")

# Spectral Centroid (right axis)
ax2 = ax1.twinx()
ax2.plot(t_centroid, centroid_n, label="Spectral Centroid (norm)", color='gray', linestyle="--")
ax2.set_ylabel("Centroid (normalized)")

# Merge legends
lines = ax1.get_lines() + ax2.get_lines()
labels = [ln.get_label() for ln in lines]
ax1.legend(lines, labels, loc="upper right")

plt.title("Normalized Timbre Features: RMS, ZCR, Spectral Centroid")
plt.tight_layout()
plt.show()

# Explicitly prevent extra figures
plt.close(fig)
```

```python
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
S_dB = librosa.power_to_db(S, ref=np.max)

plt.figure(figsize=(12,4))
librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel')
plt.colorbar(format="%+2.0f dB")
plt.title("Mel Spectrogram")
plt.tight_layout()

#plt.savefig("MelSpectogram-251206-Fame.png", dpi = 300,bbox_inches='tight')
plt.show()
```

```python

#Beat Tracking
tempo, beats = librosa.beat.beat_track(y=y, sr=sr)
beat_times = librosa.frames_to_time(beats, sr=sr)

#Plot mel spectogram with beat markers
S = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=512)
S_dB = librosa.power_to_db(S, ref=np.max)

plt.figure(figsize=(14,6))
librosa.display.specshow(S_dB, sr=sr, hop_length=512, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
tempo = float(tempo[0])
plt.title(f"Mel Spectorgram with Beat Markers (Tempo: {tempo:.1f} BPM)")

for bt in beat_times:
    plt.axvline(x=bt, color='white', linestyle='--', linewidth=1, alpha=0.8)

plt.tight_layout()

#plt.savefig("MelSpectogramwBeat-251206-Fame.png", dpi = 300,bbox_inches='tight')
plt.show()
```

```python
#onset strength envelope + onset detection

onset_env = librosa.onset.onset_strength(y=y, sr=sr)
plt.figure(figsize=(14,4))
plt.plot(librosa.times_like(onset_env),onset_env)
plt.title("Onset Strength Envelope")


#plt.savefig("OnsetStrengthEnvelope-251206-Fame.png", dpi = 300,bbox_inches='tight')
plt.show()
```

```python
# Chroma feature
chroma = librosa.feature.chroma_stft(y=y, sr=sr)

plt.figure(figsize=(12,4))
librosa.display.specshow(chroma, x_axis='time', y_axis='chroma', sr=sr)
plt.colorbar()
plt.title("Chroma Features")
plt.tight_layout()

plt.savefig("ChromaFeatures-251206-Fame.png", dpi = 300,bbox_inches='tight')
plt.show()
```
