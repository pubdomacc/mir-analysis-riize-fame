```python
#Used to define all libraries used in this code

import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Will need to iterate for each of the files inside the directory.
# Is it better to use the loop inside the plot or not
audio_path = "audio/10. Another Life.wav"

# y is waveform, sr is sampling rate.
y, sr = librosa.load(audio_path)
print("Loaded:", audio_path)

# Check what is y type, and what are their features
# Check the impact of sampling rate
print("Shape:", y.shape, "Sample rate:", sr)
```

```python
# check plt figure, and its feature

plt.figure(figsize=(12, 3))
librosa.display.waveshow(y, sr=sr)
plt.title("Waveform")
plt.tight_layout()
plt.show()
```

```python
rms = librosa.feature.rms(y=y)[0]
plt.figure(figsize=(12,3))
plt.plot(rms)
plt.title("RMS Energy")
plt.tight_layout()
plt.show()
```

```python
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
S_dB = librosa.power_to_db(S, ref=np.max)

plt.figure(figsize=(12,4))
librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel')
plt.colorbar(format="%+2.0f dB")
plt.title("Mel Spectrogram")
plt.tight_layout()
plt.show()
```

```python
chroma = librosa.feature.chroma_stft(y=y, sr=sr)

plt.figure(figsize=(12,4))
librosa.display.specshow(chroma, x_axis='time', y_axis='chroma', sr=sr)
plt.colorbar()
plt.title("Chroma Features")
plt.tight_layout()
plt.show()
```
