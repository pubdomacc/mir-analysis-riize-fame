```python
#Used to define all libraries used in this code

import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

## Feature extraction 

We compute standard DSP/ML features such as: 
- Mel-spectogram
- MFCCs
- Spectral centroid / rolloff
- Zero-crossing rate
- Chroma features

These will serve as the numerical input to the ML moderl

```python
# Will need to iterate for each of the files inside the directory.
# Is it better to use the loop inside the plot or not
audio_path = "../audio/01. Something\'s in the Water.wav"

# y is waveform, sr is sampling rate.
y, sr = librosa.load(audio_path)
print("Loaded:", audio_path)

# Check what is y type, and what are their features
# Check the impact of sampling rate
print("Shape:", y.shape, "Sample rate:", sr)
```

```python
# check plt figure, and its feature

plt.figure(figsize=(12, 3))
librosa.display.waveshow(y, sr=sr)
plt.title("Waveform")
plt.tight_layout()

#plt.savefig("Waveform-251202.png", dpi = 300,bbox_inches='tight')
plt.show()
```

```python
rms = librosa.feature.rms(y=y)[0]
plt.figure(figsize=(12,3))
plt.plot(rms)
plt.title("RMS Energy")
plt.tight_layout()

#plt.savefig("RMSEnergy-251202.png", dpi = 300,bbox_inches='tight')
plt.show()
```

```python
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
S_dB = librosa.power_to_db(S, ref=np.max)

plt.figure(figsize=(12,4))
librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel')
plt.colorbar(format="%+2.0f dB")
plt.title("Mel Spectrogram")
plt.tight_layout()

#plt.savefig("MelSpectogram-251202.png", dpi = 300,bbox_inches='tight')
plt.show()
```

```python
#Onset envelope
onset_env = librosa.onset.onset_strength(y=y, sr=sr)

plt.figure(figsize=(14,3))
plt.plot(onset_env, label='Onset Envelope')
plt.title("Onset Envelope (Energy Peaks Over Time)")
plt.xlabel("Frames")
plt.ylabel("Strength")
plt.legend()
plt.tight_layout()

plt.show()
```

```python

#Beat Tracking
tempo, beats = librosa.beat.beat_track(y=y, sr=sr)
beat_times = librosa.frames_to_time(beats, sr=sr)

#Plot mel spectogram with beat markers
S = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=512)
S_dB = librosa.power_to_db(S, ref=np.max)

plt.figure(figsize=(14,6))
librosa.display.specshow(S_dB, sr=sr, hop_length=512, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
tempo = float(tempo[0])
plt.title(f"Mel Spectorgram with Beat Markers (Tempo: {tempo:.1f} BPM)")

for bt in beat_times:
    plt.axvline(x=bt, color='white', linestyle='--', linewidth=1, alpha=0.8)

plt.tight_layout()
plt.show()
```

```python
#zero crossing Rate (ZCR)

zcr = librosa.feature.zero_crossing_rate(y,frame_length=2048, hop_length=512)[0]
plt.figure(figsize=(14,4))
plt.plot(zcr)
plt.title("Zero-Crossing Rate")
plt.show()
```

```python

hop_length = 512
n_fft = 2048
n_mels = 128

# Mel spectrogram
S = librosa.feature.melspectrogram(
    y=y,
    sr=sr,
    n_fft=n_fft,
    hop_length=hop_length,
    n_mels=n_mels
)

S_db = librosa.amplitude_to_db(S, ref=np.max)

# Mel frequency axis
mel_frequencies = librosa.mel_frequencies(n_mels=n_mels, fmin=0, fmax=sr/2)


#Spectral Centroid Overlaid on Spectogram

centroid = librosa.feature.spectral_centroid(y=y, sr=sr)[0]
times = librosa.times_like(centroid)

plt.figure(figsize=(14,4))
librosa.display.specshow(librosa.amplitude_to_db(S, ref=np.max), sr=sr, 
                         hop_length=hop_length, x_axis='time', y_axis='mel')
plt.plot(times,centroid/centroid.max()*np.max(mel_frequencies), color='w')
plt.title("Mel Spectogram + Spectral Centroid")
plt.show()



                         
```

```python
#Spectral bandwidth and spectral contrast

bandwidth = librosa.feature.spectral_bandwidth(y=y, sr=sr)[0]
contrast = librosa.feature.spectral_contrast(y=y, sr=sr)
```

```python
#onset strength envelope + onset detection

onset_env = librosa.onset.onset_strength(y=y, sr=sr)
plt.figure(figsize=(14,4))
plt.plot(librosa.times_like(onset_env),onset_env)
plt.title("Onset Strength Envelope")
plt.show()
```

```python
#Tempo curve
tempo, beats = librosa.beat.beat_track(y=y, sr=sr)

#harmonic vs percussive split(HPSS)
harmonic, percussive = librosa.effects.hpps(y)
```

```python
chroma = librosa.feature.chroma_stft(y=y, sr=sr)

plt.figure(figsize=(12,4))
librosa.display.specshow(chroma, x_axis='time', y_axis='chroma', sr=sr)
plt.colorbar()
plt.title("Chroma Features")
plt.tight_layout()

#plt.savefig("ChromaFeatures-251202.png", dpi = 300,bbox_inches='tight')
plt.show()
```

```python
print("Audio duration:", librosa.get_duration(y=y, sr=sr))
print("Last beat time:", beat_times[-1])
```
